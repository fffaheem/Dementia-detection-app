```python
def find_closest(arr,x):
    s = 1; b = -1
    for i in arr:
        d = i - x
        if d < 0:
            if s == 1:
                if d < s:
                    s = d
            else:
                if d > s:
                    s = d
        else:
            if b == -1:
                b = d
            else:
                if d < b:
                    b = d
    if b == -1:
        return s+x
    else :
        return b+x
```

```python
def csv_merger(main,secondary,csv_file_name,attribute,lower_bound,upper_bound,fid,header,full_id_row_num):
    main_data = pd.read_csv(main)
    secondary_data = pd.read_csv(secondary)

    main_data = main_data.values
    count = 0

    with open(csv_file_name, 'w', newline='') as csv_file:
        # Create a CSV writer object
        csv_writer = csv.writer(csv_file)
    
        # Write header row
        csv_writer.writerow(['MRI-ID', header])
    
    
        for row in main_data:
            main_id = row[0][:8]
            main_days = row[0][-4:]
        
            matched_rows = secondary_data.loc[secondary_data[attribute] == main_id]
            # print(matched_rows)
     
            matched = []
            for matched_row in matched_rows.values:
                secondary_id = matched_row[full_id_row_num][:8]
                secondary_days = matched_row[full_id_row_num][-4:]
        
                difference = int(secondary_days) - int(main_days)
                # print(matched_row)
                if difference < upper_bound and difference > -lower_bound:
                    matched.append(secondary_days)

            if len(matched) > 0:
                print(row[0])
                final = find_closest(map(int, matched),int(main_days))
                final_id = main_id + fid + "d" + str(final).zfill(4)
                csv_writer.writerow([row[0], final_id])
                print(final_id)
                count += 1
        
            print()
    
    
    print(str(count) +" complete")
```
